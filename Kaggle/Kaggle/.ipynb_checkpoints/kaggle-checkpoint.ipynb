# First features 

## Create features depending on bidder_id 
1. number of participated auctions for every bidder
2. number of used ips
3. number of used url
4. number of merchandises
5. number of devices 

## Use decision tree model 
1. gridSearch 
2. cross validation
3. ROC



```python
# creat train dataframe and test dataframe
trainDf = train[["bidder_id", "outcome"]]
testDf = test[["bidder_id"]]
```

```python
# add number of auctions
g = df[["bidder_id", "country"]].groupby('bidder_id').country.nunique()
def nbAuctions(row, group):
    if row["bidder_id"] in group:
        return g[row["bidder_id"]]
    else:
        return 0

trainDf["nbActions"] = trainDf.apply(lambda x : nbAuctions(x, g), axis = 1)
testDf["nbAuctions"] = testDf.apply(lambda x : nbAuctions(x, g), axis = 1)
```

```python
# add number of ip for every bidder in train dataset and test dataset
g = df[["bidder_id", "ip"]].groupby('bidder_id').ip.nunique()
def nbIps(row, group):
    if row["bidder_id"] in group:
        return g[row["bidder_id"]]
    else:
        return 0

trainDf["nbIps"] = trainDf.apply(lambda x : nbIps(x, g), axis = 1)
testDf["nbIps"] = testDf.apply(lambda x : nbIps(x, g), axis = 1)
```

```python
# add number of URL for every bidder in train dataset and test dataset
g = df[["bidder_id", "url"]].groupby('bidder_id').url.nunique()
def nbUrl(row, group):
    if row["bidder_id"] in group:
        return g[row["bidder_id"]]
    else:
        return 0

trainDf["nbUrl"] = trainDf.apply(lambda x : nbUrl(x, g), axis = 1)
testDf["nbUrl"] = testDf.apply(lambda x : nbUrl(x, g), axis = 1)

        
```

```python
# add number of merchandises for every bidder in train dataset and test dataset
g = df[["bidder_id", "merchandise"]].groupby('bidder_id').merchandise.nunique()
def nbMer(row, group):
    if row["bidder_id"] in group:
        return g[row["bidder_id"]]
    else:
        return 0

trainDf["nbMer"] = trainDf.apply(lambda x : nbMer(x, g), axis = 1)
testDf["nbMer"] = testDf.apply(lambda x : nbMer(x, g), axis = 1)
```

```python
# add number of devices for every bidder in train dataset and test dataset
g = df[["bidder_id", "device"]].groupby('bidder_id').device.nunique()
def nbDevice(row, group):
    if row["bidder_id"] in group:
        return g[row["bidder_id"]]
    else:
        return 0

trainDf["nbDevice"] = trainDf.apply(lambda x : nbDevice(x, g), axis = 1)
testDf["nbDevice"] = testDf.apply(lambda x : nbDevice(x, g), axis = 1 )
```

```python
trainDf.head()
```

```python

X = trainDf[trainDf.columns.difference(["bidder_id",'outcome'])]
y = trainDf.outcome

# split data into train data and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
    
```

```python
# import classifiers
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
```

```python
# Decision tree 
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score

param_grid = {'max_depth': np.arange(2, 10)}

tree = GridSearchCV(DecisionTreeClassifier(), param_grid)
tree.fit(X_train, y_train)
tree_preds = tree.predict_proba(X_test)[:, 1]
tree_performance = roc_auc_score(y_test, tree_preds)


print ('DecisionTree: Area under the ROC curve = {}'.format(tree_performance))
print (tree.best_params_) # {'max_depth': 9}
```

```python
# Cross validation for decisiontree max_depth = 9
num_folds = 10
seed = 0
scoring = 'f1_macro'
kfold = KFold(n_splits=num_folds, random_state=seed)
cv_result = cross_val_score(DecisionTreeClassifier(max_depth=9), X, y, cv=kfold, scoring=scoring)
print('%s: %f (%f)' % ("DecisionTreeClassifier", cv_result.mean(), cv_result.std()))
```

```python
# add number of won auction for evary bidder
g = df[["bidder_id", "time", "auction"]].groupby('auction')
def wonAuctionDict(group):
    result = {}
    for auction in group:
        row = auction[1].iloc[[-1]]
        key = row["bidder_id"].values[0]
        auctionName = row["auction"].values[0]
        if key in result.keys():
            result[key].append(auctionName)
        else:
            result[key] = [auctionName,]
    result2 = {}
    for key in result.keys():
        result2[key] = len(result[key])
    return pd.DataFrame(list(result2.items()), columns=['bidder_id', 'nbWonAuctions'])
        
            
```

```python
nbWonAuction_df = wonAuctionDict(g)
```

```python
trainDf["nbWonAuction"] = pd.merge(trainDf, nbWonAuction_df, how='left', on=['bidder_id'])["nbWonAuctions"].fillna(0)
testDf["nbWonAuction"] = pd.merge(testDf, nbWonAuction_df, how='left', on=['bidder_id'])["nbWonAuctions"].fillna(0)
```

```python
X = trainDf[trainDf.columns.difference(["bidder_id",'outcome'])]
y = trainDf.outcome

# split data into train data and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# Decision tree 
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score

param_grid = {'max_depth': np.arange(2, 10)}

tree = GridSearchCV(DecisionTreeClassifier(), param_grid)
tree.fit(X_train, y_train)
tree_preds = tree.predict_proba(X_test)[:, 1]
tree_performance = roc_auc_score(y_test, tree_preds)


print ('DecisionTree: Area under the ROC curve = {}'.format(tree_performance))
print (tree.best_params_) # {'max_depth': 9}
```

```python
X_train = trainDf.drop(['bidder_id','outcome'],axis=1)
y_train = trainDf['outcome']

rand = RandomForestClassifier(n_estimators=2000,max_depth=30,min_samples_leaf=2)
print (cross_val_score(rand,X_train,y_train,cv=10).mean())
```

```python
# Run classifier with cross-validation and plot ROC curves
from sklearn.cross_validation import StratifiedKFold
from sklearn.metrics import roc_curve, auc
from scipy import interp
X=np.array(X_train)
y=np.array(y_train)
cv = StratifiedKFold(y, n_folds=4)
classifier = RandomForestClassifier(n_estimators=2000,max_depth=25,min_samples_leaf=1)

mean_tpr = 0.0
mean_fpr = np.linspace(0, 1, 100)
all_tpr = []

for i, (tran, tet) in enumerate(cv):
    probas_ = classifier.fit(X[tran], y[tran]).predict_proba(X[tet])
    # Compute ROC curve and area the curve
    fpr, tpr, thresholds = roc_curve(y[tet], probas_[:, 1])
    mean_tpr += interp(mean_fpr, fpr, tpr)
    mean_tpr[0] = 0.0
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, lw=1, label='ROC fold %d (area = %0.2f)' % (i, roc_auc))

plt.plot([0, 1], [0, 1], '--', color=(0.6, 0.6, 0.6), label='Luck')

mean_tpr /= len(cv)
mean_tpr[-1] = 1.0
mean_auc = auc(mean_fpr, mean_tpr)
plt.plot(mean_fpr, mean_tpr, 'k--',
         label='Mean ROC (area = %0.2f)' % mean_auc, lw=2)

plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```

```python
# add time reponse to bid 
g = df[["bidder_id", "time", "auction"]].groupby('auction')
def bidderTimeReponseInOneAuction(group):
    biddersReponseTime = {} 
    i = 0
    for auction in group:
        
        last_row = None
        print (str(i)+'/'+str(len(group)))
        for _, row in auction[1].iterrows():
            
            if last_row is None:
                last_row = row
                continue
            time_difference = row['time'] - last_row['time']
            if  row["bidder_id"] in biddersReponseTime.keys():
                biddersReponseTime[row["bidder_id"]].append(time_difference)
            else :
                biddersReponseTime[row['bidder_id']] = [time_difference,]
            last_row = row
        i += 1
            
    return biddersReponseTime
            
```

```python
import pickle

def save_obj(obj, name ):
    with open( name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_obj(name ):
    with open( name + '.pkl', 'rb') as f:
        return pickle.load(f)
```

```python
result = bidderTimeReponseInOneAuction(g)
save_obj(result, "biddersReponseTime_dict")
result = load_obj("biddersReponseTime_dict")
```

```python
result_mean = {}
result_max = {}
result_min = {}
for key in result.keys():
    result_mean[key] = np.array(result[key]).mean()
    result_max[key] = np.array(result[key]).max()
    result_min[key] = np.array(result[key]).min()
        
```

```python

dic_df = pd.DataFrame(list(result_mean.items()), columns=['bidder_id', 'meanTimeReponseOneAuc'])
trainDf["meanTimeReponseOneAuc"] = pd.merge(trainDf, dic_df, how='left', on=['bidder_id'])["meanTimeReponseOneAuc"].fillna(0)
testDf["meanTimeReponseOneAuc"] = pd.merge(testDf, dic_df, how='left', on=['bidder_id'])["meanTimeReponseOneAuc"].fillna(0)

dic_df = pd.DataFrame(list(result_max.items()), columns=['bidder_id', 'maxTimeReponseOneAuc'])
trainDf["maxTimeReponseOneAuc"] = pd.merge(trainDf, dic_df, how='left', on=['bidder_id'])["maxTimeReponseOneAuc"].fillna(0)
testDf["maxTimeReponseOneAuc"] = pd.merge(testDf, dic_df, how='left', on=['bidder_id'])["maxTimeReponseOneAuc"].fillna(0)

dic_df = pd.DataFrame(list(result_min.items()), columns=['bidder_id', 'minTimeReponseOneAuc'])
trainDf["minTimeReponseOneAuc"] = pd.merge(trainDf, dic_df, how='left', on=['bidder_id'])["minTimeReponseOneAuc"].fillna(0)
testDf["minTimeReponseOneAuc"] = pd.merge(testDf, dic_df, how='left', on=['bidder_id'])["minTimeReponseOneAuc"].fillna(0)
```

# Thirds features

## Create features
1. time reponse between two own bids in one auction for every bidder (mean, max, min, std, mediam, zerosNumbers)

## Random forest

1. Cross validation
2. Roc curves

## Submission test.csv
1. create submission.csv  (kaggle score 0.88099) 

```python
# add mean time between two bids of one auction for every bidder
g = df[["bidder_id", "time", "auction"]].groupby(['bidder_id', 'auction'])
def auc_features_count(group):
    time_diff = np.ediff1d(group['time'])
    bidder_interval = {}
    if len(time_diff) == 0:
        diff_mean = 0
        diff_std = 0
        diff_median = 0
        diff_zeros = 0
        diff_min = 0
        diff_max = 0
    else:
        diff_mean = np.mean(time_diff)
        diff_std = np.std(time_diff)
        diff_median = np.median(time_diff)
        diff_max = np.max(time_diff)
        diff_min = np.min(time_diff)
        diff_zeros = time_diff.shape[0] - np.count_nonzero(time_diff)
    
    bidder_interval['tmean'] = diff_mean
    bidder_interval['tmin'] = diff_min
    bidder_interval['tmax'] = diff_max
    bidder_interval['tstd'] = diff_std
    bidder_interval['tmedian'] = diff_median
    bidder_interval['tzeros'] = diff_zeros
    bidder_interval = pd.Series(bidder_interval)
    return bidder_interval

```

```python
bidders = df.groupby('bidder_id')
bidder_inv = bidders.apply(auc_features_count)
bidder_inv = bidder_inv.reset_index()
```

```python
trainDf["tmean"] = pd.merge(trainDf, bidder_inv, how='left', on=['bidder_id'])["tmean"].fillna(0)
testDf["tmean"] = pd.merge(testDf, bidder_inv, how='left', on=['bidder_id'])["tmean"].fillna(0)

trainDf["tmax"] = pd.merge(trainDf, bidder_inv, how='left', on=['bidder_id'])["tmax"].fillna(0)
testDf["tmax"] = pd.merge(testDf, bidder_inv, how='left', on=['bidder_id'])["tmax"].fillna(0)

trainDf["tmin"] = pd.merge(trainDf, bidder_inv, how='left', on=['bidder_id'])["tmin"].fillna(0)
testDf["tmin"] = pd.merge(testDf, bidder_inv, how='left', on=['bidder_id'])["tmin"].fillna(0)

trainDf["tstd"] = pd.merge(trainDf, bidder_inv, how='left', on=['bidder_id'])["tstd"].fillna(0)
testDf["tstd"] = pd.merge(testDf, bidder_inv, how='left', on=['bidder_id'])["tstd"].fillna(0)
```

```python

trainDf["tmedian"] = pd.merge(trainDf, bidder_inv, how='left', on=['bidder_id'])["tmedian"].fillna(0)
testDf["tmedian"] = pd.merge(testDf, bidder_inv, how='left', on=['bidder_id'])["tmedian"].fillna(0)

trainDf["tzeros"] = pd.merge(trainDf, bidder_inv, how='left', on=['bidder_id'])["tzeros"].fillna(0)
testDf["tzeros"] = pd.merge(testDf, bidder_inv, how='left', on=['bidder_id'])["tzeros"].fillna(0)
```
